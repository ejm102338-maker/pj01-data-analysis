12. 욕설 필터링: 특정 단어 리스트에 사용자가 입력한 단어가 포함되어 있으면 "차단된 단어"를 출력하세요.

```python
# 변수 정의
word_list = ["하얀","걷다","뛰다"]
text = "한 소녀가 하얀 눈밭을 걸어 갑니다."

# 조건 : word_list[] in text
# 설계 : 참이면 '차단된 단어 사용', 거짓이면 '통과'

i = 0
isTrue = True

while i < len(word_list) :
    if word_list[i] in text :
        isTrue = False 
    i += 1

if isTrue :
    print("통과")
else :
    print("차단된 단어 사용")
```

17. 잔액 확인: 원하는 물건이 상점 리스트에 있고, 내 지갑의 돈이 물건 가격보다 많아야 "구매 완료"를 출력하세요.

```python
# 변수 정의
goods_list = {
            "사과":{"price":300,"size":5},
            "배": {"price":350,"size":7},
            "귤":{"price":250,"size":3},
        }

my_money = 300

my_goods = "배"

# 조건 : my_goods in goods_list 품목에 있는지 확인
# 조건 : my_money >= goods_list[my_goods]["price"]
# 설계 : 조건이 참이면 '구매완료',거짓이면 금액부족 (부족금액 표시)

if my_goods in goods_list :
    if my_money >= goods_list[my_goods]["price"]:
        print("구매완료")
    else : 
        print(f"금액부족 : { goods_list[my_goods]['price'] - my_money}원")
else :
    print("없는 물품입니다")
```
