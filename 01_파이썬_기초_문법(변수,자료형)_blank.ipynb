### (3) 연산

````python
yoga = {"A","C","D","F","G"}
health = {"A","B","F","H"}

# 교집합(AND)
set_inter = yoga & health
print("교집합 : ",set_inter)

set_inter = yoga.intersection(health)
print("교집합2 : ",set_inter)

# 합집합(OR)
set_or = yoga | health
print("합집합 : ",set_or)

set_union = yoga.union(health)
print("합집합 : ",set_or)

# 차집합
set_diff = yoga - health
print("차집합(yoga - health) : ",set_diff)
print("차집합(health - yoga) : ",health - yoga)

set_diff = yoga.difference(health)
print("차집합 : ",set_diff)

# 대칭차집합
set_s = yoga ^ health
print("대칭차집합",set_s)

````
