### 1. 미니실습

```python
html = """
<html>
  <body>
    <div id="main" class="content">
      <h1>첫 번째 헤더</h1>
      <h1 class="header">두 번째 헤더</h1>
      <p class="description">설명입니다.</p>
      <a href="https://example1.com" class="link">링크 1</a>
      <a href="https://example2.com" class="link">링크 2</a>
      <a href="https://example3.com" class="external">링크 3</a>
    </div>
    <div id="main2" class="content">
      <div class="items">
        <div>
            <img class="thumbnail" src="/data/picture1.png" />
            <div class="title">타이틀1</div>
            <div class="description">설명1</div>
        </div>
        <div>
            <img class="thumbnail" src="/data/picture2.png" />
            <div class="title">타이틀2</div>
            <div class="description">설명2</div>
        </div>
        <div>
            <img class="thumbnail" src="/data/picture3.png" />
            <div class="title test">타이틀3</div>
            <div class="description">설명3</div>
        </div>
      </div>
    </div>
  </body>
</html>
"""
```

### 1) HTML 파싱준비

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html,'html.parser')
soup
```

### 2) 추출해보기

#### soup.find("태그이름",class_=클래스이름,id=아이디이름)

```python
# 조건에 만족하는 맨 처음 요소 1개만 가져온다
# class이름이 titleㅓ이라고 써진 요소들을 가져와보자

soup.find("div",class_="title")
```

```python
# 텍스트 추출
soup.find("div",class_="title").text
```

```python
# 조건에 만족하는 모든 요소들을 가져온다
soup.find_all("div",class_="title")
```

```python
# 조건에 만족하는 모든 요소들을 가져온다

for ele in soup.find_all("div",class_="title"):
    print(ele.text)
```

### 속성찾기

```python
# img 태그를 추출해 주세요
soup.find_all("img")
for ele in soup.find_all("img"):
    print(ele.attrs)
```

```python
# a 태그를 추출해 주세요
soup.find_all("a")
```

```python
for ele in soup.find_all("a") : 
    print("attrs : ",ele.attrs["href"])
    print("attrs.get : ",ele.attrs.get("href"))
    print("get : ",ele.get("href"))
```

### 1) HTML 요청해서 불러오기

```python
# https://search.kyobobook.co.kr/robots.txt 크롤링이 가능한지 확인하는 
url = "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=kyobo"
response = requests.get(url)
print(response.status_code)
```

```python
# 누가 이 사이트에 요청을 했는가?
print(response.request.headers)
```

```python
# user-Agent 수동입력하기
headers = {
    "User-Agent" : "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Mobile Safari/537.36"
}

response = requests.get(url,headers=headers)
print(response.status_code)
```

```python
print(response.request.headers)
```

```python

# url 만들기
url = "https://search.kyobobook.co.kr/search"

keyword = "인공지능"
sort = ""

# keyword=인공지능&target=kyobo&gbCode=TOT
params = {
    "keyword" : keyword,    # 키워드
    "targete" : "kyobo",    # 교보문고 탭
    "gbCode"  : "TOT",      #
    "page"    : 1    ,      # 페이지수
    "ra"      : sort        # SORT

}
```

```python
# 요청하기
response = requests.get(url,params=params,headers=headers)
print(response.status_code)
```

```python
print(response.request.url)
```

### 2) 파싱준비

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
soup
```
