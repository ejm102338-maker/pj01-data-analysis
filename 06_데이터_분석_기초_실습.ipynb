```python
## 6) 평균체류시간(분) 의 전체 평균을 계산한 후 평균이상이면 " 체류시간 김" ,아니면 짧음

total_mean = df["평균 체류 시간(분)"].mean()
print(f"평균 체류시간은 {total_mean} 입니다")
```

### 1번째 방법

```python
# 1. 새로운 열로 만든다
df["message"] = df["평균 체류 시간(분)"] >=total_mean
df.head()
```

```python
df["message"] = df["message"].replace({True:"체류기간 김",False:"체류시간 짧음"})
df.head()
```

### 2번째 방법

```python
# np.where()
import numpy as np
df["message2"]=np.where(df["평균 체류 시간(분)"]>=total_mean,"체류시간 김","체류시간 짧음")
df.head()
```

### 3번째 방법

```python
# apply()
df["message3"]=df["평균 체류 시간(분)"].apply(lambda x : "체류시간 김" if x >= total_mean else "체류시간 짧음")
df.head()
```

# 10. 수치형 변수들의 상관관계를 히트맵으로 나타내주세요

```python
select_columns = ["월 수입(만원)", "평균 체류 시간(분)", "방문 페이지 수", "자녀 수"]

corr = df[select_columns].corr()

plt.figure(figsize=(4,3))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    vmax=1,
    vmin=-1
)
plt.title("수치형 변수들의 히트맵")
plt.show()
```
