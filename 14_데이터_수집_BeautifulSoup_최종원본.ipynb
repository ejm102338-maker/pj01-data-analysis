### User-Agent 수동 설정

```python
import requests 

url = "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=total"

response = requests.get(url)
print(response.status_code)
```

```python
# 누가 이 사이트에 요청을 했는가?
print(response.request.headers)
```

```python
# User-Agent 가 python이면 요청받은 서버는 봇이구나! 를 알게 된다.
# User-Agent 수동입력하기 : 개발자도구 > 네트워크 탭 선택 > index.html 찾기 클릭 (없으면 새로고침) > 맨 아래로 스크롤 내리면 User-Agent 있다 > 복사하기
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

url = "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=total"

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
# 누가 이 사이트에 요청을 했는가?
print(response.request.headers)
```

### HTML 가져오기

```python
# 사용자 입력
keyword = "인공지능"
```

```python
# URL 만들기
# "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=total" -> url과 query parameter를 쪼갠다. (물음표 기준)
url = "https://search.kyobobook.co.kr/search"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

# "?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&target=kyobo&gbCode=TOT&page=2"  # & 로 이어진다. 
# params로 요청하면 한글 인코딩 안해도 된다. 
# 이렇게 하면 웹 화면에서 페이지를 클릭하지 않아도 파이썬에서 우리가 원하는 웹 페이지 URL을 생성할 수 있다. 
params = {
    "keyword": keyword,        # 키워드
    "target": "kyobo",         # 교보문고 탭 선택
    "gbCode": "TOT",           # 통합검색
    "page": 1                  # 페이지 번호
}
```

## 4) 반복문으로 20개의 책 정보 수집하기

```python
# 사용자 입력
keyword = "인공지능"
```

```python
from bs4 import BeautifulSoup 
import requests 

# STEP1. URL 만들기
## Base URL
url = "https://search.kyobobook.co.kr/search"
## Headers 
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}
## Query Parameter
params = {
    "keyword": keyword,        # 키워드
    "target": "kyobo",         # 교보문고 탭 선택
    "gbCode": "TOT",           # 통합검색
    "page": 1                  # 페이지 번호
}

# STEP2. HTML 요청하기 
response = requests.get(url, params=params, headers=headers)
if response.status_code != 200:
    print("문제가 발생했습니다.") # -- 반복할 때 break 넣어야 함 

# STEP3. HTML 파싱준비
soup = BeautifulSoup(response.text, 'html.parser')

# STEP4. 책 목록 불러오기 
prod_item_list = soup.select("#shopData_list > ul > li")
print(f"{len(prod_item_list)}개의 책이 있습니다.")
```

#### 하나일 때 코드 정리
