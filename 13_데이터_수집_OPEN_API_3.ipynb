### (과제1) 연도 기간을 설정해서 반복하려면 어떻게 해아하나?

```python
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
base_url = "http://apis.data.go.kr/B552061/frequentzoneLgrViolt/getRestFrequentzoneLgrViolt"
key = os.getenv("PUBLIC_DATDA_API_KEY")

# searchYearCd = 2017 반복문
siDo = 11
guGun = 680
request_type = "json"
numOfRows = 12
pageNo = 1
```

```python
# 뭘 반복할 것인가 2015~ 2019까지
import requests
import pandas as pd

data_list = []
for searchYearCd in range(2020,2024):
    url = f"{base_url}?ServiceKey={key}&searchYearCd={searchYearCd}&siDo={siDo}&guGun={guGun}&type={request_type}&numOfRows={numOfRows}&pageNo={pageNo}"
    response = requests.get(url)

    if response.status_code != "200" : break

    data = response.json()
    item_list = data["items"]["item"]
    data_list.extend(item_list)

df = pd.DataFrame(data_list)

#df.to_csv("./data/기술문서_한국도로교통공단_법규위반별 사고다발지역 정보_년도별.hwp")
```

### (과제 2) 여러시도 데이터를 반복문으로 가져올 수 있나요?

```python
import requests
import pandas as pd
base_url = "http://apis.data.go.kr/B552061/frequentzoneLgrViolt/getRestFrequentzoneLgrViolt"
key = os.getenv("PUBLIC_DATDA_API_KEY")

searchYearCd = 2017
siDo = 11             # 반복문마다 변함
guGun = 680           # 반복문마다 변함
request_type = "json"
numOfRows = 12
pageNo = 1
```

```python
# 뭘 반복 할 것인가?
# 변수 리스트로 정의해보자
target = [(11,680),(41,135)]
target = [{"siDo":11,"guGun":680},{"siDo":41,"guGUn":135}]

data_list = []
for x in target:
    siDo = x["siDo"]
    guGun = x["guGun"]
    url = f"{base_url}?ServiceKey={key}&searchYearCd={searchYearCd}&siDo={siDo}&guGun={guGun}&type={request_type}&numOfRows={numOfRows}&pageNo={pageNo}"
    response = requests.get(url)

    if response.status_code != "200" : break

    data = response.json()
    item_list = data["items"]["item"]
    data_list.extend(item_list)
```

```python
df = pd.DataFrame(data_list)
```

```python

### 반복문을 통한 페이지 변화
pageNo = 1
data_list = []
while True:
    params["pageNo"] = pageNo
    print(f"params에 pageNo = {pageNo}가 업데이트 되었습니다.")

    response = requests.get(base_url,params=params)
    
    if response.status_code != 200 :
        print("문제가 생겼어요")
        break
    
    data = response.json()
    item_list = data["StanReginCd"][1]["row"]
    print(f"\t{len(item_list)} 개의 데이터를 수집했습니다.",end="")

    print("\t합체")

    data_list.extend(item_list)
    pageNo += 1

    if len(item_list) < 1000 : 
        print("종료")
        break
```
