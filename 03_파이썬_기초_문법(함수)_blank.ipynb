## 3) 매개변수

```python
# 덧셈 계산기 :
## 입력:a,b
## 실행: a+ b 반환

## 함수이름 : calc_add(a,b)

def calc_add(a,b):
    return a + b

result = calc_add(40,5)
print(result)
```

```python
# 사칙연산 계산기
# 입력변수 : a,b,operator
# 함수이름 : calc

def calc(a,b,oper="+"):
    if oper=="+": return a + b
    elif oper == "-": return a - b
    elif oper == "*" : return a * b
    elif oper == "/" : 
        if b == 0 :
            return "0으로 나눌 수 없습니다."
        else : 
            return a / b
    else:return "잘못입력하셨습니다."

print("+",calc(10,2))
print("-",calc(10,2,"-"))
print("*",calc(10,2,"*"))
print("/",calc(10,2,"/"))
print("/",calc(10,0.1,"/"))
print("%",calc(10,2,"%"))
```

### 심화

```python
# 사용자 정보를 입력받는 함수
# 입력이 자유롭다. name만 입력해도 되고,age만 입력해도 작동하도록 한다.

def user_info(*args):
    print(args)

user_info('홍길동')
user_info('홍길동',23)
user_info('홍길동',23,188.2)
```

```python
def user_info(**kwargs):
    print(kwargs)

user_info(name="홍길동",age=23,height="188.2")
```

## 4) lambda 함수

```python
calc_add = lambda a,b : a + b
calc_sub = lambda a,b : a - b
calc_div = lambda a,b : a / b if b != 0 else "0으로 나눌 수 없습니다."
calc_mul = lambda a,b : a * b

print(calc_add(1,2))
print(calc_sub(1,2))
print(calc_div(10,2))
print(calc_mul(10,2))
```
