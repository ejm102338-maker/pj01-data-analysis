```python
# 한번요청할때 1,000 행만 가능하다.
## 전체가 3,210... 4번요청
## startl - end
## 1 - 1000
## 1001 - 2000
## 2001 - 3000
## 3001 - 3210
```

```python
# start = 1 변수를 만든다
# 4번 반복하는 반복문을 반든다
# 반복할때 마다 end를 생성하고, start,end를출력한다 
# start, end가 반영된 url을 만든다
# url을 요청한다
# 요청ㅇ받은 데이터에서 item을 추출한다
# start를 업데이트 한다
# 반복문이 끝난 후 item_list의 갯수를 출력한다. - 3210개가 맞는지 ㅣ확ㄷ인
# 데이터프레임으로 바꾼다
# 데이터를 저장한다.

total_cnt = int(data["stationInfo"]["list_total_count"])
total_page = int(total_cnt / 1000)

if total_cnt % 1000 > 0: total_page += 1

item_list = []

#for i in range(total_page) :
for i in range(1,4000,1000) :
    if i < total_cnt :  
        #start = (i*1000) +1
        start = i-1
        end = start + 999
        #if end > total_cnt : end = total_cnt
        
        url = f"{base_url}/{key}/{request_type}/{service}/{start}/{end}"
        response = requests.get(url)
        data = response.json()
        print("받은 데이터 : ",len(data["stationInfo"]["row"]))
        end_i = 1
        for x in data["stationInfo"]["row"] : 
            item_list.append(x) 
    else : 
        break
    
    print(start,end)

    i += 1 

print(len(item_list))
```

```python
import pandas as pd

item_list_df = pd.DataFrame(item_list)
```
