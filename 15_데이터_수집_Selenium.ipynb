# 1. Chrome 창켜기

```python
from selenium import webdriver

driver = webdriver.Chrome()  # 크롬 브라우저 실행
driver.maximize_window()     # 창 최대화
```

# 2. 페이지 이동

```python
url = "https://play.google.com/store/search?hl=ko"

driver.get(url)  # 해당 URL로 이동
```

# 3. 돋보기 클릭하기

```python
from selenium.webdriver.common.by import By

btn_selector = "#kO001e > header > nav > div > div:nth-child(1) > button"
sch_btn = driver.find_element(By.CSS_SELECTOR, btn_selector)  # 돋보기 버튼 클릭
sch_btn.click()
```

# 4. 검색어 입력

```python
keyword = "배달의 민족"
input_box_selector = "#kO001e > header > nav > c-wiz > div > div > label > input"
input_box = driver.find_element(By.CSS_SELECTOR, input_box_selector)  # 검색 입력창
input_box.send_keys(keyword)  # 검색어 입력
```

# 5. 엔터 누르기

```python
from selenium.webdriver.common.keys import Keys

input_box.send_keys(Keys.ENTER)  # 엔터키 입력
```

# 9. 데이터 가져오기

```python
# 왜 내려야 하는가
# 스크롤을 내려야 댓글이 나타남
import time

display_selectore ="#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi.Vk3ZVd"
display_area = driver.find_element(By.CSS_SELECTOR, display_selectore)

rep = 5
for i in range(rep):
    display_area.send_keys(Keys.END)
    time.sleep(0.5)  # 0.5초 대기
```

# 1) 스크롤 맨 처음으로 올라가기

```python
display_area.send_keys(Keys.HOME)
```

```python
review_selector = "#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi.Vk3ZVd > div > div:nth-child(2) > div"
reviews = driver.find_elements(By.CSS_SELECTOR, review_selector)  # 리뷰 리스트 가져오기
print((len(reviews)))
```

```python
review = reviews[0]
review
```

```python
# 작성자
name = review.find_element(By.CSS_SELECTOR, ".X5PpBb").text  # 리뷰 내용만 추출
print(name)
```

```python
write_date = review.find_element(By.CSS_SELECTOR, ".bp9Aid").text  # 작성일자 추출
print(write_date)
```

```python
review_text = review.find_element(By.CSS_SELECTOR, ".h3YV2d").text  # 리뷰 내용만 추출
print(review_text)
```

```python
# 찾은 데이터 합치기
row = [{
    "name" : name,
    "write_date" : write_date,
    "review_text" : review_text
}]
```

```python
data_list = []

for i,rv in enumerate(reviews) :
    name = rv.find_element(By.CLASS_NAME, "X5PpBb").text
    write_date = rv.find_element(By.CLASS_NAME, "bp9Aid").text
    review_text = rv.find_element(By.CLASS_NAME, "h3YV2d").text
    row = {
        "name" : name,
        "write_date" : write_date,
        "review_text" : review_text
    }
    data_list.append(row)
    print(f"{i}번째 데이터가 수집되었습니다.")


```

### 4) 데이터 프레임 만들기

```python
import pandas as pd

df = pd.DataFrame(data_list)

df.head()
```

```python
df.to_csv("./data/배달의민족_구글플레이스토어_댓글수집(20260128).csv")
```
