```python
import pandas as pd

# 데이터 생성
data_dict = {
    "customer_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", None, "Eva"],
    "age": [25.0, 30.0, 28.0, 40.0, 35.0],
    "signup_date": ["2023-01-15", "2023-02-20", "2023-03-20", "2023-03-01", "2023-04-10"],
    "purchase_amount": [120.50, 120.2, 200.00, 150.75, 300.40],
    "is_active": [True, False, "yes", "no", True],
    "group": ["A", "A", "B", "C", "C"]
}

# 데이터프레임 생성
df = pd.DataFrame(data_dict)
df
```

1. is_active 열만 추출(loc 이용, iloc 이용)

```python
print(df.iloc[:,5])
print("-"*20)
print(df.loc[:,["is_active"]])
```

2. age, group 열만 추출(loc 이용, iloc 이용)

```python
print(df.loc[:,["age","group"]])
print("-"*20)
print(df.iloc[:,[2,-1]])
```

3. customer_id가 105인 행만 추출(loc 이용, iloc 이용)

```python
print(df.loc[df["customer_id"]==105,:])
print("-"*20)
print(df.loc[df.iloc[:,0]==105,:])

#print(df.iloc[df.iloc[0,"customer_id"]==105,:])
```

4. customer_id 가 103, 104, 105인 행만 추출(loc 이용, iloc 이용)

```python
print(df.loc[df["customer_id"] >= 103 , :])
print("-"*20)
print(df.loc[df.iloc[:,0]>=103 , : ])
```

5. charlie 의 group 추출(loc 이용, iloc 이용)

```python
print(df.loc[df["name"]=="Charlie","group"])
print("-"*20)
print(df.loc[df.iloc[:,1]=="Charlie","group"])
```

6. customer_id 가 102, 103인 데이터의 name, age 추출(loc 이용, iloc 이용)

```python
print(df.loc[df["customer_id"].between(102,103),["name","age"]])
print("-"*20)
print(df.loc[(df["customer_id"] == 102)|(df["customer_id"]==103),["name","age"]])
```

7. Alice 의 customer_id, name, age  추출 (loc 이용, iloc 이용)

```python
print(df.loc[df["name"]=="Alice",["customer_id","name","age"]])
print("-"*20)
print(df.loc[df.iloc[:,1]=="Alice",["customer_id","name","age"]])
```

8. Bob, Eva 의 signup_date, is_activate 추출

```python
print(df.loc[(df["name"]=="Bob")|(df["name"]=="Eva"),["name","signup_date","is_active"]])
```
