# 1. 조건문

````python
# 나이를 입력하면 미성년자인지 아닌지 구분

# 조건 나이가 19세 이상이면 성인, 미만이면 미성년자
# 변수 정의 : 
user_age = int(input("나이를 입력하세요"))

# 설계
if user_age >= 19:
    print(f"{user_age} : 성인입니다.")
else:
    print(f"{user_age} : 미성년자 입니다.")
````

````python
# 등급에 따라 메시지 출력하기
# 점수가 60점 이상이면 '합격', 그외 '불합격'

# 변수
user_point = int(input("점수를 입력하세요!"))

# 조건 : 점수가 60점 이상이면 '합격', 그외 '불합격'
# 설계 : 참이면 합격, 거짓이면 불합격
if user_point >= 60 :
    print("합격")
else : 
    print("불합격")
````

````python
# 등급에 따라 출력 (A-Perfect,B-Exellent,C-Good,그외는 '노력하세요'

#변수정의
grade = "C"

#조건
if grade == "A":
    print("Perfect")
elif grade == "B" :
    print("Exellent")
elif grade == "C" :
    print("Good")
else :
    print("노력하세요")
````

문제. 나이가 10세 이상이고, 키가 120cm 이상일 때만 놀이기구 탑승

* 탑승 가능하면 "통과" 아니면 "탑승 불가"라고 출력하기

````python
age = 10
height = 120

user_age = 20
user_height = 100

# 조건 : 나이가 10세 이상이고 키가 120 이상인경우
# 설계 : 참이면 '통과', 거짓이면 '탑승불가'
if user_age >= age and user_height >= height :
    print("통과")
else :
    print("탑승불가")

user_age = 20
user_height = 170

result = "통과" if user_age >= age and user_height >= height else "탑승불가"
print(result)
````

문제. 10만원 이상 구매하거나 "Gold" 등급이면 10% 할인

* 할인이 가능하면 "10% 할인 가능" 아니면 "10% 할인 불가" 출력

````python
# 변수정의
buy_price = 90000
buy_grade = "Silver"

# 조건 
## 조건1 : 구매금액이 10만원이상
## 조건2 : 등급이 Gold
## 조건결합 : OR
# 설계 : 참이면 '10% 할인 가능', 거짓이면 '10% 할인 불가능'
if buy_price >= 100000 or buy_grade == "Gold" : 
    print("10% 할인 가능")
else :
    print("10% 할인 불가능")

buy_price = 100000
buy_grade = "Silver"
result = "10% 할인 가능" if buy_price >= 100000 or buy_grade == "Gold" else "10% 할인 불가능"
print(result)
````

````python
# while True : , count = 5일때 멈추기, 짝수일때는 출력말기

#변수정의
count = 0

# 조건 :count ==5 ,count %2=0

while  True:
    count += 1

    #조건1
    if count % 2 == 0 : continue

    print(count)

    #조건2
    if count == 5 : break
````

````python
# 예제: 비밀번호가 맞을 때까지 입력하기

# 변수정의 : password = 1234
## 반복 : 입력(input()), 조건 : if password == "1234" 반복 끝

## 설계
while True :
    password = input("비밀번호를 입력하세요")
    if password == "1234" : 
        print("로그인 되었습니다.")
        break
    else :
        print("비밀번호를 다시 입력해 주세요")

# 변형해 보기 : 아이디, 비밀번호를 모두 설정해서 맞는지 확인하기
id = "abcd"
pwd = "123456"
count = 0

while True :
    count += 1
    # id_input = input("아이디를 입력하세요")
    # pwd_input = input("비밀번호를 입력하세요")
    txt_input = input("아이디/패스워드를 입력해 주세요")
    if "/" in txt_input :
        id_input,pwd_input = txt_input.split("/")
    else :
        print("형식에 맞지 않습니다.")
        continue

    if id_input == id and pwd == pwd_input:
        print("로그인 되었습니다.")
        break
    else :
        print("계정이 맞지 않습니다.")
    
    if count == 3 :
        print("3회이상 틀려서 5분 기다리셔야 합니다.")
        break
````

````python
# 반복문을 활용하여 "1일차 매출은 120만원입니다." 형식의 메시지를 출력하고 싶다.

# 변수정의
daily_sales = [120,150,100,120,150]

for x in range(len(daily_sales)):
    print(f"{x+1}일차 매출은 {daily_sales[x]}만원입니다")

print("="*30)

i = 0
for sales in daily_sales:
    i += 1
    print(f"{i}일차 매출은 {sales}만원입니다")
````

````python
# 반복문을 활용하여 "1일차 매출은 120만원입니다." 형식의 메시지를 출력하고 싶다.
daily_sales = [120,150,100,120,150]

for idx,daily_sale in enumerate(daily_sales):
    print(f"{idx+1}일차 매출은 {daily_sale}만원 입니다.")
````

예제 3. 고객 주문 데이터

반복문을 사용해서 총 매출을 계산하세요.

```
orders = [
    {"customer": "Kim", "amount": 120000},
    {"customer": "Lee", "amount": 85000},
    {"customer": "Park", "amount": 99000},
]
```

````python
orders = [
    {"customer": "Kim", "amount": 120000},
    {"customer": "Lee", "amount": 85000},
    {"customer": "Park", "amount": 99000},
]

tot = 0

for item in orders :
    tot += item["amount"]
    print(f"{item['customer']}님의 매출은 {item['amount']}원 입니다.")
    
print("="*30)
print(f"{tot:,}")

result = sum(idx["amount"] for idx in orders )
print(result)
````
