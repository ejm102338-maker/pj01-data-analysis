# 4. 배열 변환

```python
# 배열 변환 (sape)
data = np.arange(0,15)
print(data)

print(data.size) # 전체갯수

# sape를 변환(3,5) 
data_reshape = data.reshape(3,5)
print("data_reshape : ",data_reshape)

# 전치행렬 (행/열 전환)
data_transpose = data_reshape.T
print("data_transpose : ",data_transpose)

# n차원 배열을 1차원으로 변환 배열.flatten()
data_flatten = data_transpose.flatten()
data_reflatten = data_reshape.flatten()
print("data_flatten : ",data_flatten)
print("data_reshape : ",data_reflatten)
```
