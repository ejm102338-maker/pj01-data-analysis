16. [구형 연락처 추출] 

users = [{"name": "A", "phone": "011-1234-5678"}, {"name": "B", "phone": "010-5678-4321"}, {"name": "C", "phone": "011-9999-1231"}]

전화번호(phone)가 "011"로 시작하는 고객의 이름만 출력하세요. (문자열 .startswith() 활용해보기)

17. [재고 금액 합산]

stock = {"apple": 5, "banana": 0, "cherry": 10}

prices = {"apple": 1000, "banana": 500, "cherry": 2000}

 현재 가지고 있는 모든 과일의 총 재고 가치(수량 * 가격)를 계산하세요.

```python
stock = {"apple": 5, "banana": 0, "cherry": 10}

prices = {"apple": 1000, "banana": 500, "cherry": 2000}

result = 0
for fruit in stock :
    result += stock[fruit] * prices[fruit]
    print(f"{stock[fruit]} X {prices[fruit]} : {stock[fruit] * prices[fruit]}")    

print("="*20)
print(result)
```

19. [장바구니 결제]

cart = [{"item": "T-shirt", "price": 20000, "amount": 2}, {"item": "Pants", "price": 30000, "amount": 1}]

전체 상품의 총 결제 금액을 반복문을 통해 구하세요.

```python
cart = [{"item": "T-shirt", "price": 20000, "amount": 2}, {"item": "Pants", "price": 30000, "amount": 1}]
result = 0

for my_cart in cart :
    result += my_cart["price"] * my_cart["amount"]
    print(f"{my_cart['item']} : {my_cart['price']} X {my_cart['amount']}")

print("="*20)
print(result)
```

20. [데이터 클렌징] (isinstance() 참고) 

raw_data = [85, None, 92, "Missing", 78, None]

숫자(int) 데이터만 골라내어 새로운 리스트에 담고 그 평균을 구하세요.

```python
raw_data = [85, None, 92, "Missing", 78, None]

number_list = []
result = 0

for data in raw_data :
    if type(data) is int :
        number_list.append(data)
        result += data

print("숫자:",number_list)
print("평균:",result / len(number_list))
```
